```python
# 1. Limpiar textos con espacios extra y mayúsculas/minúsculas mezcladas
# Ejemplo con el estado de matrícula (" Regular ", "REGULAR", "regular")
df_estudiantes['estado_matricula'] = df_estudiantes['estado_matricula'].str.strip().str.upper()

# 2. Arreglar las fechas (Hay formatos mezclados como DD/MM/YYYY y YYYY-MM-DD)
df_asistencia['fecha'] = pd.to_datetime(df_asistencia['fecha'], errors='coerce')
df_calificaciones['fecha'] = pd.to_datetime(df_calificaciones['fecha'], errors='coerce')
```

deteccion de outliers y errores

```python
# 1. Forzar la conversión de texto a números. 
# errors='coerce' transformará cualquier texto raro (ej: espacios en blanco) en un valor Nulo (NaN)
df_calificaciones['nota'] = pd.to_numeric(df_calificaciones['nota'], errors='coerce')

# 2.  vemos las estadísticas descriptivas 
display(df_calificaciones['nota'].describe())

# 3. Buscamos los outliers matemáticos
outliers_notas = df_calificaciones[(df_calificaciones['nota'] < 1.0) | (df_calificaciones['nota'] > 7.0)]

print(f"Se encontraron {len(outliers_notas)} notas fuera de rango:")
display(outliers_notas)

# 4. Y hacemos el gráfico
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 4))
sns.boxplot(x=df_calificaciones['nota'], color='skyblue')
plt.title('Distribución de Notas (Búsqueda de Outliers)')
plt.show()
```

en el grafico se aprecian las notas atipicas erroneas, probablemente de tipeo del profesor

```python
# Unir estudiantes con sus inscripciones
df_master = df_estudiantes.merge(df_inscripciones, on='id_estudiante', how='inner')

# Unir lo anterior con sus notas
df_master = df_master.merge(df_calificaciones, on='id_inscripcion', how='left')
#al usar how='left' me aseguro de no perder a los estudiantes que quizás están inscritos pero aún no tienen ninguna nota registrada.
```

```python
# 1. Limpieza de texto en la columna 'carrera' (Estandarización)
# - str.strip(): Quita los espacios al principio y al final
# - str.title(): Pone la primera letra en mayúscula y el resto en minúscula
df_master['carrera'] = df_master['carrera'].str.strip().str.title()

# 2. Filtrar los Outliers (Dejar solo las notas reales entre 1.0 y 7.0)
# Así evitamos que la nota "82.1" arruine nuestro cálculo
df_master_limpio = df_master[(df_master['nota'] >= 1.0) & (df_master['nota'] <= 7.0)]

# 3. Volver a agrupar y calcular el promedio
promedio_carreras = df_master_limpio.groupby('carrera')['nota'].mean().sort_values()

print("=== Promedio Real por Carrera (Datos Limpios) ===")
display(promedio_carreras.round(2))
```
